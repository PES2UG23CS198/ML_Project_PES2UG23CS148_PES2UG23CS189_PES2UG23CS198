```python
import pandas as pd
import numpy as np

# Load cleaned dataset
df = pd.read_csv("../data/listings_clean_v1.csv")
df.head()
```

```python
df = df.drop(columns=["id", "name", "host_name"])
```

```python
# Price per minimum night (helps detect very expensive short stays)
df["price_per_night"] = df["price"] / df["minimum_nights"]

# Total potential reviews (reviews * min_nights)
df["reviews_ratio"] = df["number_of_reviews"] / (df["minimum_nights"] + 1)

# Combine location info: (optional, for geo analysis)
df["lat_long"] = df["latitude"].astype(str) + "_" + df["longitude"].astype(str)
```

```python
# One-hot encode room_type
df = pd.get_dummies(df, columns=["room_type", "neighbourhood"], drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

num_cols = ["price", "minimum_nights", "number_of_reviews", "price_per_night", "reviews_ratio"]
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
df.to_csv("../data/listings_features_v1.csv", index=False)
print("Feature engineered dataset saved!")
```
